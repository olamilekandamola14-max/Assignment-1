```python
# Q3: BMI Calculator
Given a weight of 70 kg and a height of 1.75 meters:

Calculate the BMI using the formula: bmi = weight / (height  2).

Print the BMI rounded to 1 decimal place.

Use if-elif-else to print the correct category:

BMI < 18.5 ➡️ "Underweight"

BMI < 25 ➡️ "Normal weight"

BMI < 30 ➡️ "Overweight"

BMI >= 30 ➡️ "Obese"
```

```python
# ANSWER
weight=70
height=1.75

# using the formula;bmi = weight / (height**2)

BMI = weight/height**2
print(f"your bmi is {BMI:.2f}")
# Using Conditional Statement

if BMI>=30:
    print("Obese")
elif BMI>=25:
    print("Over weight")
elif BMI>=18.5:
    print("Normal weight")
else:
    print("Under weight")

```

```python
# Q6: Store Discount System
A store offers discounts based on membership type ("premium" or "regular") and purchase amount. Write a program to calculate the final price:

Premium Members: 20% discount if purchase is $100 or more; 10% discount if under $100.

Regular Members: 10% discount if purchase is $100 or more; 5% discount if under $100.

Others: 0% discount.
Test your code with member_type = "premium" and purchase_amount = 150.
```

```python
# ANSWER
member_type="premium"
purchase_amount = 150

if member_type=="premium":
    if purchase_amount>=100:
        discount=0.2
    else:
        discount=0.1
elif member_type=="regular":
    if purchase_amount>=100:
        discount=0.1
    else:
        discount=0.05
else:
    discount=0

discount_amount=purchase_amount*discount

final_price =purchase_amount-discount_amount

print(f"final_price for {member_type} with goods of {purchase_amount} is {final_price}")

```

```python
# Q7: Movie Ticket Pricing (Exercise 2.3)
Write a program that determines movie ticket prices using nested logic:

Ask the user for their Age.

Ask the user if today is Tuesday (Yes/No).

Calculate the ticket price based on these rules:

Standard price by age: Adult (18+) = $15, Senior (65+) = $10, Child (<18) = $8.

Special Rule: If it is Tuesday, everyone pays $7 regardless of age.
```

```python
# ANSWER
age=int(input("How old are you"))
day=input("is today Tuesday?")

if day=="Yes":
    price=7
    print("everyone pay $",price)
else:
    if age >=65:
        price=10
    elif age >=18:
        price=15
    else:
        price=8

    print("your ticket price is $",price)

```
